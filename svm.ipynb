```python

from sklearn.svm import SVC
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python

iris = datasets.load_iris()
```

```python

X = iris.data[:, :2]
y = iris.target
```

```python

modelsvm = SVC(kernel='linear')
modelsvm.fit(X, y)
```

```python

predictions = modelsvm.predict(X)
```

```python

accuracy = modelsvm.score(X, y)
print("Accuracy of SVM:", accuracy)
```

```python

def plot_decision_boundary(X, y, model):
    h = .02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', s=50)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title('SVM Decision Boundary for Iris Dataset')
    plt.colorbar()
    plt.show()
```

```python

plot_decision_boundary(X, y, modelsvm)
```

```python

new_samples = np.array([[5, 3], [6, 2.5]])
predictions = modelsvm.predict(new_samples)
print("Predictions for new samples:", predictions)
```
